```python
%load_ext autoreload
%autoreload 2
import sys
import numpy as np
```

```python
sys.path.append("../")
from libintroia.logistic_regression import LogisticRegression
from libintroia.dataset import DataXY
```

## Regresión logística

```python
DATASET_FILENAME="data/clase_6_dataset.txt" 
data = DataXY(DATASET_FILENAME)
train_ds, test_ds = data.split()
```

```python
train_ds.shape, test_ds.shape
```

```python
n_feat = train_ds.shape[1]-1
```

```python
x_train = train_ds[:,:n_feat]
y_train = train_ds[:,n_feat].reshape(-1,1)
x_test = test_ds[:,:n_feat]
y_test = test_ds[:,n_feat].reshape(-1,1)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
model = LogisticRegression()
_ = model.fit(x_train,y_train)
```

### Evaluación del modelo

```python
from libintroia.utils import plot_confusion_matrix
from libintroia.metrics import Precision, Accuracy, Recall

metrics = [ Precision(), Accuracy(), Recall() ]
```

#### Training set

```python
predicted = model.predict(x_train,y_train)
actual = y_train
plot_confusion_matrix(predicted, actual, "Training set")
print("Precision:", metrics[0](actual,predicted))
print("Accuracy:", metrics[1](actual,predicted))
print("Recall:", metrics[2](actual,predicted))
```

#### Test set

```python
predicted = model.predict(x_test,y_test)
actual = y_test
plot_confusion_matrix(predicted, actual, "Test set")
print("Precision:", metrics[0](actual,predicted))
print("Accuracy:", metrics[1](actual,predicted))
print("Recall:", metrics[2](actual,predicted))
```
