## 4. SVD

#### 4.1 Ejercicio

Dada:
$
A = 
\left[\begin{matrix}1 & 0 & 1\\0 & 1 & 1\end{matrix}\right]
$. Halle sin usar un software de cálculo su descomposición en valores singulares.

##### Desarrollo

Se busca obtener una matriz de 2x2, por lo tanto se calcula:

$$
A A^T = 
\left[\begin{matrix}2 & 1\\1 & 2\end{matrix}\right]
$$

Paso 1. Se obtienen los autovalores de $A.A^T$:

$$
det(A A^T - \lambda I) = 0
$$

$$
det(
\left[\begin{matrix}2 & 1\\1 & 2\end{matrix}\right]
-
\left[\begin{matrix}-\lambda & 0\\0 & -\lambda\end{matrix}\right]
) = 0
$$

$$
det(\left[\begin{matrix}2 - \lambda & 1\\1 & 2 - \lambda\end{matrix}\right]) = 0
$$

$$
\lambda^2 - 4 \lambda + 3 = 0 
$$

$$
\therefore \lambda_1 = 3, \lambda_2 = 1
$$

Paso 2. Construcción de matriz $\Sigma$ de misma dimensión que $A A^T$.

$$
\lambda_i = \sigma_i^2 \therefore \sigma_i=\sqrt{\lambda_i}
$$

$$
\sigma_1 = \sqrt{3}
$$

$$
\sigma_2 = 1
$$

$$
\Sigma = 
\left[\begin{matrix}\sigma_1 & 0 & 0\\0 & \sigma_2 & 0\end{matrix}\right]
=
\left[\begin{matrix}\sqrt{3} & 0 & 0\\0 & 1 & 0\end{matrix}\right]
$$



Paso 3. Se obtienen los autovectores de $A.A^T$

Para $\lambda_1=3$:

$$
(AA^T-\lambda_1I)X=0
$$

$$
\left[\begin{matrix}2 - \lambda_{1} & 1\\1 & 2 - \lambda_{1}\end{matrix}\right]
\left[\begin{matrix}x_{1}\\x_{2}\end{matrix}\right] = 0
$$

$$
\left[\begin{matrix}2 - 3 & 1\\1 & 2 -  3\end{matrix}\right]
\left[\begin{matrix}x_{1}\\x_{2}\end{matrix}\right] = 0
$$

$$
\therefore
x_1 = x_2
$$

$$
u_1 = 
\begin{bmatrix}
1 \\
1 
\end{bmatrix}
$$

Para $\lambda_2=1$:


$$
(A A^T-\lambda_2I)X=0
$$

$$
\left[\begin{matrix}2 - \lambda_{2} & 1\\1 & 2 - \lambda_{2}\end{matrix}\right]
\left[\begin{matrix}x_{1}\\x_{2}\end{matrix}\right] = 0
$$

$$
\left[\begin{matrix}2 - 1 & 1\\1 & 2 -  1\end{matrix}\right]
\left[\begin{matrix}x_{1}\\x_{2}\end{matrix}\right] = 0
$$

$$
\therefore
x_1 = -x_2
$$

$$
u_2 = 
\begin{bmatrix}
-1 \\
1 
\end{bmatrix}
$$


Paso 4. Construcción de $U$ a partir de autovectores $u_1$ y $u_2$ normalizados:

$$
U = 
\begin{bmatrix}
\hat{u_1} && \hat{u_2} \\
\end{bmatrix}
=
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 && -1 \\
1 && 1 \\
\end{bmatrix}
$$

Paso 5. Calcular V.

$$
V = 
\begin{bmatrix}
v_1 && v_2 && v_3\\
\end{bmatrix}
$$

dónde:

$$
v_1 
= \frac{1}{\sigma_i}A^Tu_1 
= \frac{\sqrt{3}}{3}
  \left[\begin{matrix}1 & 0\\0 & 1\\1 & 1\end{matrix}\right]
  \left[\begin{matrix}\frac{\sqrt{2}}{2}\\\frac{\sqrt{2}}{2}\end{matrix}\right]
= \left[\begin{matrix}\frac{\sqrt{6}}{6}\\\frac{\sqrt{6}}{6}\\\frac{\sqrt{6}}{3}\end{matrix}\right]  
$$


$$
v_2 
= \frac{1}{\sigma_2}A^Tu_2 
= 1
\left[\begin{matrix}1 & 0\\0 & 1\\1 & 1\end{matrix}\right]
\left[\begin{matrix}- \frac{\sqrt{2}}{2}\\\frac{\sqrt{2}}{2}\end{matrix}\right]
=
\left[\begin{matrix}- \frac{\sqrt{2}}{2}\\ \frac{\sqrt{2}}{2}\\0\end{matrix}\right]
$$

y:

$$
v_3 = v_1  \times v_2 
= 
\left[\begin{matrix}\frac{\sqrt{6}}{6}\\\frac{\sqrt{6}}{6}\\\frac{\sqrt{6}}{3}\end{matrix}\right]  
\times 
\left[\begin{matrix}- \frac{\sqrt{2}}{2}\\ \frac{\sqrt{2}}{2}\\0\end{matrix}\right]
= \left[\begin{matrix}- \frac{\sqrt{3}}{3}\\- \frac{\sqrt{3}}{3}\\\frac{\sqrt{3}}{3}\end{matrix}\right]
$$


Finalmente:

$$
A = U \Sigma V^T
$$

siendo:

$$
U = \left[\begin{matrix}\frac{\sqrt{2}}{2} & - \frac{\sqrt{2}}{2}\\\frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2}\end{matrix}\right]
$$

$$
\Sigma = \left[\begin{matrix}\sqrt{3} & 0 & 0\\0 & 1 & 0\end{matrix}\right]
$$

$$
V = 
\left[\begin{matrix}\frac{\sqrt{6}}{6} & - \frac{\sqrt{2}}{2} & - \frac{\sqrt{3}}{3}\\\frac{\sqrt{6}}{6} & \frac{\sqrt{2}}{2} & - \frac{\sqrt{3}}{3}\\\frac{\sqrt{6}}{3} & 0 & \frac{\sqrt{3}}{3}\end{matrix}\right]
$$

#### 4.2 Ejercicio

Dada:
$
A = 
\left[
\begin{matrix}
1 & 0\\
1 & 0\\
1 & 1\\
\end{matrix}
\right]
$. Halle sin usar un software de cálculo su descomposición en valores singulares.

##### Desarrollo

Se busca obtener una matriz de 2x2, por lo tanto se calcula:

$$
A^TA = 
\left[\begin{matrix}3 & 1\\1 & 1\end{matrix}\right]
$$

Paso 1. Se obtienen los autovalores de $A^TA$:

$$
det(A^T A - \lambda I) = 0
$$

$$
det(
\left[
\begin{matrix}
3 & 1\\
1 & 1
\end{matrix}\right]
-
\left[\begin{matrix}-\lambda & 0\\0 & -\lambda\end{matrix}\right]
) = 0
$$

$$
det(
\left[
\begin{matrix}3 - \lambda & 1\\1 & 1 - \lambda\end{matrix}\right]) = 0
$$

$$
\lambda^2 - 4 \lambda + 2 = 0 
$$

$$
\therefore \lambda_1 = 2+\sqrt{2}, \lambda_2 = 2-\sqrt{2}
$$

Paso 2. Construcción de matriz $\Sigma$ de misma dimensión que $A^TA$.

$$
\sigma_1 = \sqrt{\sqrt{2} + 2}
$$

$$
\sigma_2 = \sqrt{\sqrt{2} - 2}
$$

$$
\Sigma = 
\left[
\begin{matrix}
\sigma_1 & 0        \\ 
0        & \sigma_2 \\ 
0        & 0 \\ 
\end{matrix}\right]
=
\left[
\begin{matrix}
\sqrt{\sqrt{2} + 2} & 0 \\
0 & \sqrt{\sqrt{2} - 2} \\
0 & 0\end{matrix}\right]
$$



Paso 3. Se obtienen los autovectores de $A^TA$

Para $\lambda_1=\sqrt{\sqrt{2} + 2}$:

$$
(A^TA-\lambda_1I)X=0
$$

$$
\left[
\begin{matrix}
3 - \lambda_{1} & 1\\
1 & 1 - \lambda_{1}
\end{matrix}\right]
\left[\begin{matrix}x_{1}\\x_{2}\end{matrix}\right] = 0
$$

$$
\left[
\begin{matrix}
3 - \sqrt{\sqrt{2} + 2} 1\\
1 & 1 -  \sqrt{\sqrt{2} + 2}\end{matrix}\right]
\left[\begin{matrix}x_{1}\\x_{2}\end{matrix}\right] = 0
$$

$$
\therefore
x_2 = - \frac{1}{1 - \sqrt{2}} x_1
$$

$$
v_1 = 
\begin{bmatrix}
- \frac{1}{1 - \sqrt{2}} \\
1 
\end{bmatrix}
$$

Para $\lambda_2=\sqrt{\sqrt{2} - 2}$:


$$
(A^TA-\lambda_2I)X=0
$$

$$
\left[
\begin{matrix}
3 - \lambda_{2} & 1\\
1 & 1 - \lambda_{2}\end{matrix}\right]
\left[\begin{matrix}x_{1}\\x_{2}\end{matrix}\right] = 0
$$

$$
\left[
\begin{matrix}
3 - \sqrt{\sqrt{2} - 2} & 1\\
1 & 1 -  \sqrt{\sqrt{2} - 2}\end{matrix}\right]
\left[\begin{matrix}x_{1}\\x_{2}\end{matrix}\right] = 0
$$

$$
\therefore
x_2 = - \frac{1}{1 + \sqrt{2}}x_1
$$

$$
v_2 = 
\begin{bmatrix}
- \frac{1}{1 + \sqrt{2}} \\
1 
\end{bmatrix}
$$

Paso 4. Construcción de $V$ a partir de autovectores $v_1$ y $v_2$ normalizados:

$$
V = 
\begin{bmatrix}
\hat{v_1} && \hat{v_2} \\
\end{bmatrix}
=
\begin{bmatrix}
\frac{\sqrt{2}}{2 \sqrt{2 - \sqrt{2}}} & - \frac{\sqrt{2}}{2 \sqrt{\sqrt{2} + 2}} \\
\frac{\sqrt{2 - \sqrt{2}}}{2}          & \frac{\sqrt{\sqrt{2} + 2}}{2}
\end{bmatrix}
$$

Paso 5. Calcular U.

$$
U = 
\begin{bmatrix}
u_1 && u_2 && u_3\\
\end{bmatrix}
$$

dónde:

$$
u_1 
= \frac{1}{\sigma_1}Av_1 
=
\frac{1}{\sqrt{\sqrt{2} + 2}}
\left[\begin{matrix}1 & 0\\1 & 0\\1 & 1\end{matrix}\right]
\left[\begin{matrix}- \frac{1}{1 - \sqrt{2}}\\1\end{matrix}\right]
=
\left[\begin{matrix}\frac{1}{2}\\\frac{1}{2}\\\frac{\sqrt{2}}{2}\end{matrix}\right]
$$

$$
u_2 
= \frac{1}{\sigma_2}Av_2
=
\frac{1}{\sqrt{2 - \sqrt{2}}}
\left[\begin{matrix}1 & 0\\1 & 0\\1 & 1\end{matrix}\right]
\left[\begin{matrix}- \frac{1}{1 + \sqrt{2}}\\1\end{matrix}\right]
=
\left[\begin{matrix}- \frac{1}{2}\\- \frac{1}{2}\\\frac{\sqrt{2}}{2}\end{matrix}\right]
$$

y:

$$
u_3 = u_1  \times u_2 
= 
\left[\begin{matrix}\frac{1}{2}\\\frac{1}{2}\\\frac{\sqrt{2}}{2}\end{matrix}\right]
\times
\left[\begin{matrix}- \frac{1}{2}\\- \frac{1}{2}\\\frac{\sqrt{2}}{2}\end{matrix}\right]
=
\left[\begin{matrix}\frac{\sqrt{2}}{2}\\- \frac{\sqrt{2}}{2}\\0\end{matrix}\right]
$$


Finalmente

$$
A = U \Sigma V^T
$$

siendo:

$$
U = 
\left[\begin{matrix}\frac{1}{2} & - \frac{1}{2} & \frac{\sqrt{2}}{2}\\\frac{1}{2} & - \frac{1}{2} & - \frac{\sqrt{2}}{2}\\\frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2} & 0\end{matrix}\right]
$$

$$
\Sigma = 
\left[\begin{matrix}\sqrt{\sqrt{2} + 2} & 0\\0 & \sqrt{2 - \sqrt{2}}\\0 & 0\end{matrix}\right]
$$

$$
V = 
\left[\begin{matrix}\frac{\sqrt{2}}{2 \sqrt{2 - \sqrt{2}}} & - \frac{\sqrt{2}}{2 \sqrt{\sqrt{2} + 2}}\\\frac{\sqrt{2 - \sqrt{2}}}{2} & \frac{\sqrt{\sqrt{2} + 2}}{2}\end{matrix}\right]
$$
